```python
"""
refrence 
gc 
stack and heap 

. How does Python memory management work?
‚û°Ô∏è Stack vs Heap
‚û°Ô∏è Reference counting
‚û°Ô∏è Garbage collection

"""

'''
Python manages memory automatically through a private heap and two primary reclamation strategies: reference counting and a generational garbage collector. 
Stack vs. Heap
Python divides memory into two distinct regions to handle different types of data: 

Stack Memory: Stores temporary data related to function execution, including function call 
frames, local variables, and parameters. It follows a Last-In-First-Out (LIFO) order; memory 
is allocated when a function is called and immediately reclaimed when it returns.

Heap Memory: A large, unorganized pool where all Python objects (lists, dicts, strings, etc.) 
are stored. Variables on the stack are merely "names" or pointers that reference the actual
 data living on the heap. 

Reference Counting
This is Python's primary "first line of defense" for memory management. 
How it works: Every object on the heap maintains a counter of how many active references point to it.
Increment/Decrement: The count increases when you assign an object to a new variable or add it to a container (like a list); 
it decreases when a variable goes out of scope or is explicitly deleted with del.

Immediate Cleanup: As soon as an object's reference count hits zero, Python immediately deallocates it. 

Garbage Collection (GC)
Reference counting alone cannot handle circular references (e.g., Object A points to Object B, and Object B points back to A), as their counts never reach zero. To solve this, Python uses a Generational Garbage Collector: 
Cycle Detection: It periodically scans objects to find and break these unreachable cycles.
Three Generations: Objects are categorized into Generations 0, 1, and 2 based on their survival.
Gen 0: Newly created objects; scanned most frequently.
Gen 1 & 2: Objects that survive a collection cycle are "promoted" to an older generation, which is scanned less often.
Manual Control: You can use the gc module to manually trigger collection via gc.collect() or adjust threshold settings. 





'''
```
