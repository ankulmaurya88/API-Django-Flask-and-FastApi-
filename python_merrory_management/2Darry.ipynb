```python
def genretor(n):
    for i in range (n):
        yield i

for i in genretor(10):
    print(i)
```

```python
def genretor1(n):
    for i in range (n):
        yield i
        yield i+1
        yield i
        yield i+1

for i in genretor1(10):
    print(i)
```

```python
def g(n):
    for i in range(n):
        yield i
        yield from [i+1, i+2]
```

```python
for i in g(10):
    print(i,end=" ")
    
```

```python
def sub(i):
    yield i+1
    yield i+2
    return 999
```

```python
def g(n):
    for i in range(n):
        yield i
        result = yield from sub(i)
        print("Returned:", result)

for i in g(10):
    print(i,end=" ")
```

```python
def sub():
    yield 1
    return 50

def main():
    result = yield from sub()
    print(result)
```

```python
for i in main():
    print(i)
```
