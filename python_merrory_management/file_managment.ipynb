```python
""" 
1. Basic Modes (w, r, a)
These are the most common modes. They allow for only one type of operation (either reading or writing).
Mode	Name	Behavior	Pointer Position
r	Read	Opens a file for reading. Fails if the file doesn't exist.	Start of file.
w	Write	Creates a new file or overwrites an existing one (wipes it).	Start of file.
a	Append	Adds data to the end of the file. Creates it if it doesn't exist.	End of file.

"""


"""
2. Extended Modes (r+, w+, a+)
The + symbol adds the "missing" capability. These are Update Modes, meaning you can both Read AND Write.
Mode	Name	Behavior
r+	Read + Write	Can read and write. Does not wipe the file, but writing will overwrite existing text character-by-character from the start.
w+	Write + Read	Can read and write. Wipes the file completely as soon as it opens. Good for creating a file and testing it immediately.
a+	Append + Read	Can read and write. Does not wipe the file. All writes go to the end, regardless of where the pointer is.

"""

# 1. Handling Large Files (Memory Efficiency)
# Interviewer's often ask: "How would you read a 10GB file if you only have 8GB of RAM?"

```

```python
from fpdf import FPDF

pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)
pdf.cell(200, 10, txt="Hello World!", ln=True, align='C')
pdf.output("file.pdf")
```

```python
from docx import Document

doc = Document()
doc.add_heading('My Document', 0)
doc.add_paragraph('This is a Word file saved via Python.')
doc.save('file.docx')
```
