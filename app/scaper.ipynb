```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.w3schools.com/"

response=requests.get(url)
```

```python
if response.status_code == 200:
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    print("Page title:", soup.title.string)

    # Save just the text to a .txt file
    with open("page_content.txt", "w", encoding="utf-8") as file:
        file.write(soup.get_text(separator="\n", strip=True))
    print("Text content saved successfully.")
```

```python
if response.status_code == 200:
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Save the prettified HTML to a .html file
    with open("page_source.html", "w", encoding="utf-8") as file:
        file.write(soup.prettify())
    print("HTML source saved successfully.")
```
